## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python

```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    # Determine BMI category based on the calculated BMI
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
def main():
    try:
        weight = float(input("Enter your weight in kg: "))
        height = float(input("Enter your height in meters: "))
        bmi = calculate_bmi(weight, height)
        category = bmi_category(bmi)
        print(f"Your BMI is: {bmi}")
        print(f"You are classified as: {category}")
    except ValueError as e:
        print(f"Error: {e}")

main()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
import ipywidgets as widgets
from scipy import stats

def plot_with_slope(slope_factor):
    # Generate random data
    np.random.seed(42)
    x = np.random.normal(0, 2, 100)
    y = slope_factor * x + np.random.normal(0, 1.5, 100)
    
    # Calculate trend line
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    line = slope * x + intercept
    
    # Create plot
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(x, y, c=y, cmap='viridis', alpha=0.6, label='Data Points')
    plt.plot(x, line, color='red', label=f'Trend Line (R = {r_value**2:.3f})')
    
    # Add labels and styling
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.title(f'Scatter Plot with Slope Factor: {slope_factor}')
    plt.colorbar(scatter, label='Y Value Magnitude')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Create interactive slider
interactive_plot = interactive(
    plot_with_slope, 
    slope_factor=widgets.FloatSlider(
        min=-5, 
        max=5, 
        step=0.1, 
        value=2,
        description='Slope:'
    )
)

# Display the interactive plot
interactive_plot
````

````python
from fastapi import FastAPI, HTTPException, Depends, status
from fastapi.security import OAuth2PasswordBearer, OAuth2PasswordRequestForm
from pydantic import BaseModel
# Import List and Optional types from typing module
# List is used for declaring arrays/lists of specific types (e.g., List[Patient])
# Optional is used to declare that a field can be None (e.g., medical_history field)
from typing import List, Optional
import jwt
from datetime import datetime, timedelta

# Initialize FastAPI app
app = FastAPI(title="Patient Management System API")

# Secret key for JWT token
SECRET_KEY = "your-secret-key"
ALGORITHM = "HS256"

# OAuth2 scheme
oauth2_scheme = OAuth2PasswordBearer(tokenUrl="token")

# Data Models
class PatientBase(BaseModel):
    name: str
    age: int
    condition: str
    medical_history: Optional[str] = None

class PatientCreate(PatientBase):
    pass

class Patient(PatientBase):
    id: int
    created_at: datetime

    class Config:
        from_attributes = True

# Mock database
patients_db = {}
user_db = {"admin": {"username": "admin", "password": "admin123"}}

# Authentication functions
def create_access_token(data: dict):
    to_encode = data.copy()
    expire = datetime.utcnow() + timedelta(minutes=30)
    to_encode.update({"exp": expire})
    encoded_jwt = jwt.encode(to_encode, SECRET_KEY, algorithm=ALGORITHM)
    return encoded_jwt

async def get_current_user(token: str = Depends(oauth2_scheme)):
    try:
        payload = jwt.decode(token, SECRET_KEY, algorithms=[ALGORITHM])
        username: str = payload.get("sub")
        if username is None:
            raise HTTPException(status_code=401, detail="Invalid authentication credentials")
    except jwt.PyJWTError:
        raise HTTPException(status_code=401, detail="Invalid authentication credentials")
    return username

# API Endpoints
@app.post("/token")
async def login(form_data: OAuth2PasswordRequestForm = Depends()):
    user = user_db.get(form_data.username)
    if not user or form_data.password != user["password"]:
        raise HTTPException(
            status_code=status.HTTP_401_UNAUTHORIZED,
            detail="Incorrect username or password"
        )
    access_token = create_access_token(data={"sub": form_data.username})
    return {"access_token": access_token, "token_type": "bearer"}

@app.post("/patients/", response_model=Patient)
async def create_patient(patient: PatientCreate, username: str = Depends(get_current_user)):
    patient_id = len(patients_db) + 1
    patient_dict = patient.dict()
    patient_dict.update({
        "id": patient_id,
        "created_at": datetime.now()
    })
    patients_db[patient_id] = patient_dict
    return patient_dict

````

###  Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt

def process_medical_image(image: np.ndarray) -> np.ndarray:
    """
    Processes a medical image by normalizing and applying a Gaussian filter.

    Args:
        image (np.ndarray): Input image as a numpy array.

    Returns:
        np.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a valid numpy array.
    """
    try:
        # Validate input
        if not isinstance(image, np.ndarray):
            raise ValueError("Input must be a numpy array.")

        # Normalize the image (scale pixel values to range [0, 1])
        normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))

        # Apply a Gaussian filter for smoothing
        filtered_image = gaussian_filter(normalized_image, sigma=1)

        return filtered_image

    except Exception as e:
        print(f"An error occurred while processing the image: {e}")
        raise

# Example usage
# Example medical image as a numpy array
image = np.random.randint(0, 255, (256, 256), dtype=np.uint8)

# Process the image
processed_image = process_medical_image(image)

# Display the original and processed images
plt.figure(figsize=(14, 6))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Medical Image")
plt.axis('on')
plt.xticks(np.linspace(0, 200, num=5))  # Set x-axis scale from 0 to 200
plt.yticks(np.linspace(0, 200, num=5))  # Set y-axis scale from 0 to 200

# Processed image
plt.subplot(1, 2, 2)
img = plt.imshow(processed_image, cmap='gray')
plt.title("Processed Medical Image")
plt.axis('on')
plt.xticks(np.linspace(0, 200, num=5))  # Set x-axis scale from 0 to 200
plt.yticks(np.linspace(0, 200, num=5))  # Set y-axis scale from 0 to 200

# Add a color bar for the processed image
cbar = plt.colorbar(img, ax=plt.gca(), orientation='vertical', fraction=0.046, pad=0.04)
cbar.set_label("Intensity (Normalized)")

plt.tight_layout()
plt.show()

````


###  Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
from typing import List, Optional
from pydantic import BaseModel, ValidationError
import statistics

# Define a Patient model using Pydantic for data validation
class Patient(BaseModel):
    id: int
    name: str
    age: int
    condition: str
    medical_history: Optional[str] = None

class PatientManager:
    def __init__(self):
        # Initialize an empty dictionary to store patient records
        self.patients = {}

    def add_patient(self, patient_data: dict) -> str:
        """
        Adds a new patient record after validating the data.

        Args:
            patient_data (dict): Dictionary containing patient details.

        Returns:
            str: Success message.
        """
        try:
            # Validate and create a Patient instance
            patient = Patient(**patient_data)
            if patient.id in self.patients:
                return f"Patient with ID {patient.id} already exists."
            self.patients[patient.id] = patient
            return f"Patient {patient.name} added successfully."
        except ValidationError as e:
            return f"Validation Error: {e}"

    def update_patient(self, patient_id: int, updated_data: dict) -> str:
        """
        Updates an existing patient record.

        Args:
            patient_id (int): ID of the patient to update.
            updated_data (dict): Dictionary containing updated patient details.

        Returns:
            str: Success or error message.
        """
        if patient_id not in self.patients:
            return f"Patient with ID {patient_id} not found."
        try:
            # Validate and update the patient record
            updated_patient = Patient(**{**self.patients[patient_id].model_dump(), **updated_data})
            self.patients[patient_id] = updated_patient
            return f"Patient {updated_patient.name} updated successfully."
        except ValidationError as e:
            return f"Validation Error: {e}"

    def get_patient(self, patient_id: int) -> Optional[Patient]:
        """
        Retrieves a patient record by ID.

        Args:
            patient_id (int): ID of the patient to retrieve.

        Returns:
            Optional[Patient]: The patient record or None if not found.
        """
        return self.patients.get(patient_id, None)

    def get_all_patients(self) -> List[Patient]:
        """
        Retrieves all patient records.

        Returns:
            List[Patient]: List of all patient records.
        """
        return list(self.patients.values())

    def get_average_age(self) -> float:
        """
        Calculates the average age of all patients.

        Returns:
            float: The average age of patients.
        """
        if not self.patients:
            return 0.0
        ages = [patient.age for patient in self.patients.values()]
        return statistics.mean(ages)

    def get_condition_statistics(self) -> dict:
        """
        Calculates the frequency of each condition among patients.

        Returns:
            dict: A dictionary with conditions as keys and their counts as values.
        """
        condition_counts = {}
        for patient in self.patients.values():
            condition_counts[patient.condition] = condition_counts.get(patient.condition, 0) + 1
        return condition_counts

# Example usage
manager = PatientManager()

# Add patients
print(manager.add_patient({"id": 1, "name": "John Doe", "age": 30, "condition": "Flu"}))
print(manager.add_patient({"id": 2, "name": "Jane Smith", "age": 25, "condition": "Cold"}))

# Update a patient
print(manager.update_patient(1, {"age": 31}))

# Retrieve a patient
patient = manager.get_patient(1)
print(patient)

# Get all patients
print(manager.get_all_patients())

# Get average age
print(f"Average Age: {manager.get_average_age()}")

# Get condition statistics
print(f"Condition Statistics: {manager.get_condition_statistics()}")
````

###  Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Fake imported data from the "PatientManager" class
data = [
    {"id": 1, "name": "John Doe", "age": 31, "condition": "Flu", "medical_history": None},
    {"id": 2, "name": "Jane Smith", "age": 25, "condition": "Cold", "medical_history": None},
    {"id": 3, "name": "Alice Johnson", "age": 40, "condition": "Healthy", "medical_history": "Diabetes"},
    {"id": 4, "name": "Bob Brown", "age": 35, "condition": "Flu", "medical_history": "Asthma"},
    {"id": 5, "name": "Charlie White", "age": 50, "condition": "Healthy", "medical_history": None},
]

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Function to display a bar chart of age distribution
def plot_age_distribution():
    plt.figure(figsize=(8, 5))
    plt.bar(df["name"], df["age"], color="skyblue")
    plt.title("Age Distribution")
    plt.xlabel("Name")
    plt.ylabel("Age")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

# Function to display a pie chart of conditions
def plot_condition_distribution():
    condition_counts = df["condition"].value_counts()
    plt.figure(figsize=(8, 5))
    plt.pie(condition_counts, labels=condition_counts.index, autopct="%1.1f%%", startangle=140, colors=["lightgreen", "lightblue", "salmon"])
    plt.title("Condition Distribution")
    plt.show()

# Function to display a scatter plot of age vs. condition
def plot_age_vs_condition():
    plt.figure(figsize=(8, 5))
    plt.scatter(df["condition"], df["age"], c="orange", alpha=0.7)
    plt.title("Age vs. Condition")
    plt.xlabel("Condition")
    plt.ylabel("Age")
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()

# Function to filter data by condition and display a summary
def filter_by_condition(condition):
    filtered_df = df[df["condition"] == condition]
    if filtered_df.empty:
        print(f"No data available for condition: {condition}")
    else:
        print(filtered_df)
        filtered_df.plot(kind="bar", x="name", y="age", figsize=(10, 6), title=f"Data for Condition: {condition}")
        plt.show()

# Interactive dashboard
def dashboard():
    print("Data Visualization Dashboard")
    print("1. Age Distribution")
    print("2. Condition Distribution")
    print("3. Age vs. Condition")
    print("4. Filter by Condition")
    
    # Interactive elements
    interact(filter_by_condition, condition=widgets.Dropdown(options=df["condition"].unique(), description="Condition:"))
    print("\nUse the dropdown above to filter data by condition.")
    print("\nRun the following functions to view specific plots:")
    print(" - plot_age_distribution()")
    print(" - plot_condition_distribution()")
    print(" - plot_age_vs_condition()")

# Run the dashboard
dashboard()

# Note: The above code is a simplified example and may require additional error handling and validation in a production environment.
# Ensure you have the required libraries installed
# pip install pandas matplotlib ipywidgets
# To run the interactive dashboard, use Jupyter Notebook or Jupyter Lab
# Ensure you have the required libraries installed
````
